```python
def find_all_routes(graph, start_line, end_line, current_route, all_routes):
    if start_line == end_line:
        all_routes.append(current_route[:])  # Add a copy of the current route to all_routes
    else:
        for destination, station in graph[start_line]:
            if destination not in [line for line, _ in current_route]:
                current_route.append([destination, station])
                find_all_routes(graph, destination, end_line, current_route, all_routes)
                current_route.pop()  # Backtrack by removing the last line

def find_routes_between_lines(graph, start_line, end_line):
    all_routes = []  # Store all possible routes
    current_route = [[start_line, '']]  # Initialize the current route with the starting line
    find_all_routes(graph, start_line, end_line, current_route, all_routes)
    return all_routes

def filter_routes(routes, end_line):
    return [route for route in routes if route[-1][0] == end_line]

def filter_final_routes(filtered_routes):
  master_list = []
  for item in filtered_routes:
    line_list = []
    switch_list = []
    for r in item:
      line_list.append(r[0])
      if r[1]=='':
        continue
      else:
        switch_list.append(r[1])
    master_list.append([line_list,switch_list])
  final_list = []
  for item in master_list:
      if len(item[1])==len(set(item[1])):
        final_list.append(item)
      else:
        continue
  return final_list


possible_routes = [['Line 1','Line 5'],['Line 2','Line 4']]

master_list = {}
for item in possible_routes:
  start_line = item[0]
  end_line = item[1]
  routes = find_routes_between_lines(lines_data, start_line, end_line)
  filtered_routes = filter_routes(routes, end_line)
  ans = filter_final_routes(filtered_routes)
  master_list[tuple(item)]=ans

for item in master_list:
      print(item,"\n")
      print(master_list[item],"\n")
```
